```python
def sum_sequence(sequence):
  result = 0
  for number in sequence:
    result += number  # result = result + number

  return result

sum_sequence(range(10))
```

```python
def prod_sequence(sequence):
  result = 1
  for number in sequence:
    result *= number
  return result

prod_sequence(range(1, 5))
```

```python
def mean_sequence(sequence):
    return sum_sequence(sequence) / len(sequence)

mean_sequence(range(10))
```

```python
def longest_string(list_of_strings: list[str]) -> str:
    return max(list_of_strings, key=len)

longest_string(["c", "bb", "aaa"]) == "aaa"
```

```python
TENDERS_DATA_URL = (
    "https://github.com/astrojuanlu/ie-mbd-python-data-analysis-i/"
    "raw/main/data/tenders.es.json"
)
```

```python
import pandas as pd
import requests

df = pd.DataFrame(requests.get(TENDERS_DATA_URL).json()["data"])
df.head()
```

```python
def clean_tenders_data(df):
    df = df.copy()
    df["type_slug"] = df["type"].str.get("slug")
    df["awarded_date"] = df["awarded"].str[0].str.get("date")
    df["purcharser_id"] = df["purchaser"].str.get("id").astype(int)
    df["deadline_length_days"] = df["deadline_length_days"].astype(float)
    return df.drop(columns=["type", "awarded", "purchaser"]).set_index("id")
```
