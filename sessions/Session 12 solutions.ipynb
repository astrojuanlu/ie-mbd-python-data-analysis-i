```python
LOBBY_DATA_URL = (
    "https://github.com/astrojuanlu/ie-mbd-python-data-analysis-i/raw/main/data/commission-lobbists-meetings.csv"
)
```

```python
df_ec = pd.read_csv(LOBBY_DATA_URL)
df_ec.head()
```

```python
LOBBY_DATA_AI_GEN_URL = (
    "https://github.com/astrojuanlu/ie-mbd-python-data-analysis-i/raw/main/data/"
    "lobby_data_ai.csv"
)
```

```python
df_lobby_data = pd.read_csv(LOBBY_DATA_AI_GEN_URL)
df_lobby_data.head(1)
```

```python
df_ec_augmented = df_ec.merge(
    df_lobby_data, how="left", left_on="lobbyist", right_on="Entity"
)
```

```python
(
    df_ec.set_index("lobbyist")
    .join(
        df_lobby_data.set_index("Entity"),
        how="left",
    )
)
```
