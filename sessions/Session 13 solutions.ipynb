```python
import pandas as pd
```

```python
LOBBY_DATA_URL = (
    "https://github.com/astrojuanlu/ie-mbd-python-data-analysis-i/raw/main/data/commission-lobbists-meetings.csv"
)
```

```python
df_ec = pd.read_csv(LOBBY_DATA_URL)
df_ec.head()
```

```python
LOBBY_DATA_AI_GEN_URL = (
    "https://github.com/astrojuanlu/ie-mbd-python-data-analysis-i/raw/main/data/"
    "lobby_data_ai.csv"
)
```

```python
df_lobby_data = pd.read_csv(LOBBY_DATA_AI_GEN_URL)
df_lobby_data.head(1)
```

```python
df_ec_augmented = df_ec.merge(
    df_lobby_data, how="left", left_on="lobbyist", right_on="Entity"
)
```

```python
df_ec_augmented.loc[df_ec_augmented["location"].isna()]
```

```python
# decision: fill the only missing value
df_ec_augmented = df_ec_augmented.fillna({"location": "UNKNOWN"})
df_ec_augmented.loc[6900]
```

```python
# decision: drop the columns
df_ec_augmented = df_ec_augmented.drop(columns=["Unnamed: 8", "Unnamed: 9", "Unnamed: 10"])
df_ec_augmented.head(2)
```
