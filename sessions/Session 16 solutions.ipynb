```python
TENDERS_DATA_URL = (
    "https://github.com/astrojuanlu/ie-mbd-python-data-analysis-i/"
    "raw/main/data/tenders.es.json"
)
```

```python
import requests
```

```python
data_raw = requests.get(TENDERS_DATA_URL).json()
tenders = data_raw["data"]
```

```python
df_tenders = pd.DataFrame(tenders)
df_tenders.head(1)
```

```python
df_tenders["category"].nunique()
```

```python
unique_categories = set()
for tender in tenders:
  unique_categories.add(tender["category"])

print(len(unique_categories))
```

```python
df_tenders["deadline_length_days"].astype(float).mean()
```

```python
import statistics
```

```python
deadlines = []
for tender in tenders:
  if "deadline_length_days" in tender:
    deadlines.append(float(tender["deadline_length_days"]))

statistics.mean(deadlines)
```

```python
df_tenders["category"].value_counts()
```

```python
categories_data = {}
for category in unique_categories:
    categories_data[category] = 0

for tender in tenders:
    categories_data[tender["category"]] += 1

categories_data
```
